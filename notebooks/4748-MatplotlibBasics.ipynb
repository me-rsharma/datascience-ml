## Introduction to Matplotlib

Visualizing data is crucial in data science as helps to quickly understand the trends & relationship in your dataset.
**Matplotlib** is one of the most popular library for plotting with Python.Matplotlib is known as the '**Grandfather**' of plotting and visualization libraries for Python and the reason is that other plotting or visualization librabries in Python are built on the top of Matplotlib such as **seaborn**, pandas built-in visualization

**Matplotlib** is heavily inspired by the plotting function of the **MatLab** programming language

```python
import matplotlib.pyplot as plt
```
