### Handle **HTML** tables data using Pandas

Pandas helps to read data from HTML tables having tag **<table>**
<span style="background-color:red; color:white; padding:2px">IMPORTANT</span> Not every table on a website is wrap under HTML **table** tag

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/World_population'
```

```python
table = pd.read_html(url)
```
