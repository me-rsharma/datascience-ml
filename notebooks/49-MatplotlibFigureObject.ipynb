### Understanding Matplotlib Figure Object

#### PART ONE: Figure Object
So far we have discussed about the functional API of Matplotlib to plot and visialize the data. But there is a comprehensive way of plotting and visualizing data using OOP API through **Figure** Object

The idea behond using Figure Object is to create a figure object and then add axes to the figure object and then plot on those axes

<span style='background-color:red; padding:2px; color:white'>NOTE</span>: The Figure object we're about to show is technically not visisble unless we add axes to it

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creat Figure object
plt.figure()
```

```python
# we can see above Figure object has 0 Axes
# add axes
fig = plt.figure()
fig.add_axes([0,0, 1, 1]) # x-axis, y-axis, x-width, y-width
```

```python
fig = plt.figure()
fig.add_axes([0,0, 1, 1]) # x-axis, y-axis, x-width, y-width
fig.add_axes([0.5,0.5, 1, 1]) # x-axis, y-axis, x-width, y-width
```

```python
fig = plt.figure() # canvas
fig.add_axes([0,0, 1,1]) # full width of canvas
# half of the width of canvas
fig.add_axes([0,0,0.5,0.5])
```

```python
fig = plt.figure() # canvas
fig.add_axes([0, 0, 1, 1])
fig.add_axes([0.2,0.2, 0.5, 0.5])
```

```python
# plot on axes
x = np.linspace(start=0, stop=1, num=26)
y = np.random.randint(low=0.5, high=26, size=16)
```

```python
fig = plt.figure()
axes_one = fig.add_axes([0,0,1,1])
axes_one.plot(x)
axes_two = fig.add_axes([0.5,0.5, 1,1])
axes_two.plot(y)
```
